```python
import numpy as np
import pandas as pd
```

```python
#Clase para dar formato a los strings
class color:
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
#Leemos los datos
dataRaw = pd.read_csv('tennis.csv')
dataRaw
```

```python
#Cogemos al cabecera de la tabla
head = dataRaw.columns
```

```python
#Pasamos los datos a un array
data = np.asarray(dataRaw)
```

```python
#Función para calcular la entropia.
#Recibe un array con los valores de la clase
#Devuelve un float.
def entropy(data):
    entropy = 0
    length = len(data)
    
    dif = np.unique(data, return_counts=True)

    for num in dif[1]:
        p = num/length
        entropy += (p * np.log2(1/p))
    
    return entropy
```

```python
#Función para calcular la ganancia de un atributo dentro de un conjunto de datos
#Recibe un conjunto de datos y un número que indica el atributo que se desea
#Devuelve un float
def gain(data,att):
    gain = 0
    att = data[:,att]
    h = entropy(data[:,-1])
    
    dif = np.unique(att, return_counts=True)

    for i in range(len(dif[0])):
        c = dif[1][i]
        a = dif[0][i]
        length = len(att)
        
        gain += c/length*entropy(data[np.where(att==a),-1][0])
        
    return h - gain
```

```python
solution = np.array([])
numDec = 4

for i in range(len(head)-1):
    print(color.BOLD + "Ganancia de", head[i] + color.END)
    solution = np.append(solution, gain(data,i))
    
    print(np.round(solution[-1],numDec))
    print("")
    
print("El atributo " + color.BOLD,
      #Nombre del atributo con mayor ganancia
      head[np.argmax(solution)]
      ,color.END + " es el que mayor ganancia tiene. " + color.BOLD + "(",
      #Ganancia de dicho atributo
      np.round(np.amax(solution),numDec)
      ,")" + color.END, sep='')
```
