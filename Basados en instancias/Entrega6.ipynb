```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score as ac
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Clase para dar formato a los strings
class color:
   BOLD = '\033[1m'
   UND = '\033[4m'
   END = '\033[0m'
```

```python
#Método para generar un clasificador KNeighbor con un k determinado
#y los datos pasados en X (Entrenamiento) e Y (test)
def KNeighbors(X, Y, k):

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/3)

    #print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
    
    #Clasificador
    clf = KNeighborsClassifier(algorithm='brute',n_neighbors=k)
    
    #Entrenamiento
    clf.fit(X_train,Y_train)
    
    #Predicción
    Y_pred = clf.predict(X_test)
    
    accuracy = ac(Y_test,Y_pred)
    
    return accuracy

    
```

```python
#Método para conseguir el mejor valor de k en función de la tasa de error para cada pareja de números
def repeated(array):
    
    #Lista de únicos
    uniques = np.empty((0,4))
    
    for e in array:

        if(uniques.shape[0] == 0):
            
            uniques = np.append(uniques,[e],axis=0)
            continue
            
        flag = 0
        for u in uniques:
            
            #Comprobamos que no se encuentra ya en la lista de únicos
            if((e[1]==u[1]) & (e[2]==u[2])):
                flag = 1
                
                #Si se encuentra y tine una tasa de error mejor entonces actualizamos la k y la tasa
                if(e[0] > u[0]):
                    u[0] = e[0]
                    u[3] = e[3]
                break
                
        #Si no estaba en la lista de únicos lo añadimos
        if(flag==0):
            uniques = np.append(uniques,[e],axis=0)
 
    for u in uniques:
        print(color.BOLD + "\n El mejor valor de k para ",
              int(u[1]), " y ",int(u[2]), " es k = ", int(u[3]), color.END, sep='')
```

```python
#Método encargado de llamar al método que genera el clasificador para cada 
def applyKNN(X,Y):
    
    bestResults = np.empty((0,4))
    
    for k in range(1,8):
        print("\n" + color.BOLD + color.UND
              + "Para el valor k = ", k, " estas son las parejas de números que mejor se diferencian: "
              + color.END +"\n", sep='')

        bestAcc = 1
        accuracys = np.empty((0,4))
        results = np.empty((0,4))
    
        #Todas las posibles parejas de números entre 5 y 9
        for i in range(5,10):
            for j in range(i+1,10):
                index = np.where((Y == i) | (Y == j))
                #print(index[0])
                Xsub = np.delete(X,index,0)
                Ysub = np.delete(Y,index,0)
                #print(i," ",j)
                #print(Xsub.shape)

                accuracy = KNeighbors(Xsub,Ysub,k)

                accuracys = np.append(accuracys,[[accuracy, i, j, k]],axis=0)
                
        #Nos quedamos con la pareja que tenga la mínima tasa de error (Máxima precisión)
        maxValue = np.amax(accuracys, axis=0)[0]
        indexs = np.argwhere(accuracys == maxValue)

        for index in indexs:
            results = np.append(results,[accuracys[index[0]]],axis=0)

        for result in results:
            bestResults = np.append(bestResults,[result],axis=0)
            print(color.BOLD, int(result[1]), " y ", int(result[2]), " (Tasa de error del ",
                  #La tasa de error es 1 - tasa de acierto
                  round(((1 - result[0])),6)
                  , ")" + color.END, sep='')
    
    return bestResults
```

```python
dataRaw = pd.read_csv('digits_from_5_to_9.csv')
dataRaw
```

### Parte I

```python
#Lo transformamos en un array
data = np.asarray(dataRaw)

X,Y = data[:,:-1], data[:,-1]

results = applyKNN(X,Y)
```

### Por tanto aquí tenemos los mejores valores de k para cada pareja

```python
repeated(results)
```
